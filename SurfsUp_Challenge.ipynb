```python
# Dependencies
import numpy as np
import datetime as dt
import pandas as pd
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june0 = dt.date(2010, 6, 1)
july0 = dt.date(2010, 7, 1)
june1 = dt.date(2011, 6, 1)
july1 = dt.date(2011, 7, 1)
june2 = dt.date(2012, 6, 1)
july2 = dt.date(2012, 7, 1)
june3 = dt.date(2013, 6, 1)
july3 = dt.date(2013, 7, 1)
june4 = dt.date(2014, 6, 1)
july4 = dt.date(2014, 7, 1)
june5 = dt.date(2015, 6, 1)
july5 = dt.date(2015, 7, 1)
june6 = dt.date(2016, 6, 1)
july6 = dt.date(2016, 7, 1)
june7 = dt.date(2017, 6, 1)
july7 = dt.date(2017, 7, 1)

results = []
results = session.query(Measurement.date, Measurement.tobs)
results = session.query(Measurement.date, Measurement.tobs).\
filter((Measurement.date >= june0) & (Measurement.date < july0)|
       (Measurement.date >= june1) & (Measurement.date < july1)|
       (Measurement.date >= june2) & (Measurement.date < july2)|
       (Measurement.date >= june3) & (Measurement.date < july3)|
       (Measurement.date >= june4) & (Measurement.date < july4)|
       (Measurement.date >= june5) & (Measurement.date < july5)|
       (Measurement.date >= june6) & (Measurement.date < july6)|
       (Measurement.date >= june7) & (Measurement.date < july7)).all()
```

```python
#  3. Convert the June temperatures to a list.
june_temperature = list(results)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temperature_df = pd.DataFrame(results, columns=['date', 'June Temps'])
june_temperature_df = june_temperature_df.set_index(june_temperature_df['date'], inplace=False)
print(june_temperature_df)
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.
june_temperature_df.describe()
```
